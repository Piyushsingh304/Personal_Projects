## **Introduction:**

Hyperspectral imaging involves capturing detailed spectral information from each pixel in an image, providing rich data across multiple bands. This makes hyperspectral images particularly valuable in applications such as remote sensing, environmental monitoring, and agriculture. In this project, we explore the use of 3D Convolutional Neural Networks (CNNs) for feature extraction from hyperspectral data, focusing on how varying filter sizes in the convolutional layers influence the network's ability to extract meaningful features.

We use the Indian Pines hyperspectral dataset, which contains multiple spectral bands for each pixel in a region of interest. The goal of this analysis is to assess the impact of different 3D convolution filter sizes on the extracted feature maps and to understand the characteristics of the resulting activations. By visualizing and comparing the feature maps generated by various filter sizes, we can gain insights into how the network interprets the hyperspectral data at different scales. Additionally, we analyze key metrics, including the mean activation, standard deviation, and sparsity of the feature maps, to quantify the effects of filter size on the model's learning process.

This approach helps demonstrate the potential of 3D CNNs in extracting spatial-spectral features from hyperspectral images, contributing to a deeper understanding of how network architecture influences performance in such high-dimensional datasets.


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv3D, MaxPooling3D, Flatten, Dense
import matplotlib.pyplot as plt

# Load the hyperspectral dataset
hyperspectral_image = np.load('/content/indianpinearray.npy')

# Visualize a single band of the hyperspectral image
plt.figure(figsize=(6, 6))
plt.imshow(hyperspectral_image[:, :, -1], cmap='viridis')
plt.title("Single Band of Hyperspectral Image")
plt.axis('off')
plt.show()

# Define a function to create a 3D CNN model with a specific filter size
def create_3d_cnn(input_shape, filter_size):
    inputs = Input(shape=input_shape)

    # 3D Convolution and MaxPooling layers
    x = Conv3D(filters=32, kernel_size=filter_size, activation='relu', padding='same', name="conv1")(inputs)
    x = MaxPooling3D(pool_size=(2, 2, 2), name="pool1")(x)

    x = Conv3D(filters=64, kernel_size=filter_size, activation='relu', padding='same', name="conv2")(x)
    x = MaxPooling3D(pool_size=(2, 2, 2), name="pool2")(x)

    x = Conv3D(filters=128, kernel_size=filter_size, activation='relu', padding='same', name="conv3")(x)
    x = MaxPooling3D(pool_size=(2, 2, 2), name="pool3")(x)

    # Flattening and dense layers for feature extraction
    feature_map = Flatten(name="flatten")(x)
    feature_map = Dense(256, activation='relu', name="dense")(feature_map)

    model = Model(inputs=inputs, outputs=feature_map)
    return model

# Prepare the data
input_shape = (hyperspectral_image.shape[0], hyperspectral_image.shape[1], hyperspectral_image.shape[2], 1)
input_data = np.expand_dims(hyperspectral_image, axis=0)  # Add batch dimension
input_data = np.expand_dims(input_data, axis=-1)  # Add channel dimension

# Filter sizes for comparison
filter_sizes = [(3, 3, 3), (5, 5, 5), (3, 3, 7)]
feature_maps = {}
intermediate_feature_maps = {}

for filter_size in filter_sizes:
    # Create model
    model = create_3d_cnn(input_shape, filter_size)
    model.summary()

    # Extract intermediate feature maps before flattening
    intermediate_model = Model(inputs=model.input, outputs=model.get_layer("pool3").output)
    intermediate_features = intermediate_model.predict(input_data)
    intermediate_feature_maps[filter_size] = intermediate_features

    # Flattened feature maps
    features = model.predict(input_data)
    feature_maps[filter_size] = features

    # Visualize intermediate feature maps
    plt.figure(figsize=(12, 6))
    # Extract a 2D slice or use an aggregated projection
    slice_to_visualize = intermediate_features[0, :, :, 0, 0]  # Adjust indices if necessary
    plt.imshow(slice_to_visualize, cmap='viridis')
    plt.title(f"Feature Map Slice with Filter Size {filter_size}")
    plt.axis('off')
    plt.show()

    # Visualize flattened features
    plt.figure(figsize=(12, 4))
    plt.imshow(features.reshape(1, -1), cmap='viridis', aspect='auto')
    plt.title(f"Flattened Features with Filter Size {filter_size}")
    plt.axis('off')
    plt.show()

# Activation metrics analysis
metrics = {'filter_size': [], 'mean_activation': [], 'std_activation': [], 'sparsity': []}

for filter_size, features in intermediate_feature_maps.items():
    # Calculate metrics
    mean_activation = np.mean(features)  # Overall mean activation
    std_activation = np.std(features)  # Overall standard deviation
    sparsity = np.sum(np.abs(features) < 1e-3) / features.size  # Sparsity across all activations

    # Store metrics
    metrics['filter_size'].append(filter_size)
    metrics['mean_activation'].append(mean_activation)
    metrics['std_activation'].append(std_activation)
    metrics['sparsity'].append(sparsity)

    print(f"Metrics for filter size {filter_size}:")
    print(f"  Mean Activation: {mean_activation:.4f}")
    print(f"  Standard Deviation: {std_activation:.4f}")
    print(f"  Sparsity: {sparsity:.4f}")

# Plot the metrics
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.bar([str(f) for f in metrics['filter_size']], metrics['mean_activation'], color='skyblue')
plt.title("Mean Activation")

plt.subplot(1, 3, 2)
plt.bar([str(f) for f in metrics['filter_size']], metrics['std_activation'], color='lightgreen')
plt.title("Standard Deviation of Activation")

plt.subplot(1, 3, 3)
plt.bar([str(f) for f in metrics['filter_size']], metrics['sparsity'], color='salmon')
plt.title("Sparsity")

plt.suptitle("Comparison of Feature Map Metrics for Different Filter Sizes")
plt.show()
```
